## Looking at df an its columns it is observed that , we do not have explicite user details that voted for movie (we have avarage voting) so we should be using collaborative filtering as we can not create user-movie matrix to review our data.

## Hence we would go for content based approach. Lets use Genres, avrage vote and langauge as features or contents of it

```python
df_selective=df[['id', 'title', 'genres', 'original_language', 'vote_average', 'vote_count']]
df_selective.head(3)
```

# Genre is in json form. lets convert it in list

```python
import json

# Function to convert JSON string to list of genre names
def extract_genres(json_str):
    genres_list= json.loads(json_str)
    return [genre['name'] for genre in genres_list]

# Apply the function to the 'Genres' column
df_selective['genres'] = df_selective['genres'].apply(extract_genres)
```

## We shall consider movies with certain number of votes, this to avoid processing  less voted movies. Lets decide that certain number of threshold value using graph

```python
# Create a scatter plot for vote_count

plt.figure(figsize=(10, 6))
plt.scatter(df_selective['id'], df_selective['vote_count'], color='blue', alpha=0.6)

# Add title and labels
plt.title('Scatter Plot of Vote Counts for Movies', fontsize=16)
plt.xlabel('Movie ID', fontsize=14)
plt.ylabel('Vote Count', fontsize=14)

# Show the plot
plt.show()
```

## Graph is dense at bottom. Assumeing its 100. So we shall take threshold value as 100 and consider those movies only which number ofr votes are minimum 100

```python
df_selective = df_selective[df_selective['vote_count'] >= 100]
df_selective.head()
```

```python
df_selective.shape
```

```python
#  Convert Genere from list format to single string form. This is necessary for further processing
def preprocess_genres(genres):
    return ' '.join(genres)

# Apply preprocessing
df_selective['genres'] = df_selective['genres'].apply(preprocess_genres)

```

## The Genre in text from to be converted into numerics. We will make use of tokenizers


```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize the CountVectorizer
cv = CountVectorizer()

# Convert the genres column into a matrix of token counts
genre_matrix = cv.fit_transform(df_selective['genres'])
```

```python
type(genre_matrix)
```

## Similary we shall update the Langauge columns

```python
# Initialize the CountVectorizer
cv1 = CountVectorizer()

langauge_matrix = cv1.fit_transform(df_selective['original_language'])
```

```python
type(langauge_matrix)
```

# Now lets deal with other column of " avg vote" and combine it with above matrices and form single dataframe

```python
from sklearn.preprocessing import StandardScaler

scaled_votes = StandardScaler().fit_transform(df_selective['vote_average'].values.reshape(-1, 1))

##reshape(-1,1) means create n=len(arr) arrays each having 1 element in each
```

## Now lets create Dataframe again with these scaled and matrices

```python
# Convert scaled_features to DataFrame
scaled_votes_df = pd.DataFrame(scaled_votes, index=df_selective['id'])

# Convert langauge_matrix to DataFrame
langauge_matrix_df = pd.DataFrame(scaled_votes, index=df_selective['id'])

# Convert genre_matrix to DataFrame
genre_matrix_df = pd.DataFrame(genre_matrix.toarray(), index=df_selective['id'])

# Combine numerical features and genre features
combined_features = pd.concat([scaled_votes_df , genre_matrix_df,langauge_matrix_df], axis=1)
```

```python
combined_features.head(3)
```

## NOW We have data in required format. Ie each movie ID has its associated feature (Genre, langaauge and avg vote ) in numeric format. Lets calculate similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Compute the similarity matrix
similarity_matrix = cosine_similarity(combined_features)
similarity_df = pd.DataFrame(similarity_matrix, index=df_selective['id'], columns=df_selective['id'])
```

```python
similarity_df.head()
```

## Recommendation Function

```python
def recommend_movies(movie, n_neighbors=5):
  #Get the id  of the movie name
  movie_id=df_selective.loc[df_selective['title']==movie,'id'].values

  # Get the similarity scores for the movie
  similar_movies = similarity_df[movie_id[0]].sort_values(ascending=False)

  # Exclude the movie itself
  similar_movies = similar_movies.drop(movie_id[0])

  # Get the indexes/id of top N similar movies
  recommended_movie_ids = similar_movies.head(n_neighbors).index

  movie_list = df_selective.loc[df['id'].isin(recommended_movie_ids), 'title'].tolist()


  return movie_list

```
